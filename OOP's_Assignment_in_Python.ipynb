# **Python OOPs Questions**

1. What is Object-Oriented Programming (OOP)?
- Object-Oriented Programming (OOP) is a programming approach that organizes code around objects and data rather than just functions and logic. Instead of writing one big block of code, you break the program into small, reusable objects — just like the real world


2. What is a class in OOP+
- A class is a blueprint or template used to create objects. It defines what data (attributes) an object will hold and what actions (methods) it can perform. By itself, a class doesn't occupy memory — memory is only allocated when an object is created from it.

3. What is an object in OOP?

- An object is a real instance created from a class — it actually exists in memory and can be used in your program. Every object has three things: an identity (unique name), a state (data/values it holds), and behavior (actions it can perform).



4. What is the difference between abstraction and encapsulation?
- The Core Difference
Abstraction → Hides what is unnecessary — shows only what the user needs to see

​
Encapsulation → Hides how it works internally — bundles data and methods together with controlled access


5.  What are dunder methods in Python+
- Dunder methods (short for Double UNDERscore) are special built-in methods in Python that start and end with double underscores like __init__, __str__, __add__. They are also called magic methods because Python calls them automatically behind the scenes — you never call them directly.

6. Explain the concept of inheritance in OOP
- Inheritance is the ability of one class to acquire properties and methods of another class — so you don't have to rewrite the same code again. The class that gives its properties is called the Parent (Base) class, and the class that receives them is called the Child (Derived) class.

7. What is polymorphism in OOP?
- The word polymorphism comes from Greek — "poly" = many and "morph" = forms. In OOP, it means the same method or function behaves differently depending on the object that calls it. Simply put — one name, many behaviors

8. How is encapsulation achieved in Python?
- Encapsulation in Python is achieved by restricting direct access to data using naming conventions with underscores, and controlling access through getter and setter methods. Python doesn't have strict access modifiers like Java or C++, but uses underscore conventions instead.



9. What is a constructor in Python?
- A constructor is a special method that runs automatically when an object is created from a class. Its main job is to initialize the object's attributes right at the time of creation — so the object is ready to use immediately. In Python, the constructor is the __init__() method.

10.  What are class and static methods in Python?
- The Class Method and Static Method
   - Class Method
A class method is bound to the class itself, not the object. It takes cls as the first parameter and can read or modify class-level data — changes apply to all objects of that class.
  - Static Method
A static method knows nothing about the class or its objects. It has no self or cls — it's basically a regular utility function that just logically belongs inside the class.
  


11.  What is method overloading in Python?
- Method Overloading means defining multiple methods with the same name but with different parameters (number or type). However, Python does NOT support true method overloading natively — if you define the same method twice, Python simply replaces the first one with the second.

12. What is method overriding in OOP?
- Method Overriding is when a child class redefines a method that already exists in the parent class — with the same name and same parameters. The child class provides its own version of the method, replacing the parent's behavior when called on a child object.

13. What is a property decorator in Python?
- Property is a built-in decorator in Python that lets you access a method like an attribute — no parentheses needed. It's the clean, Pythonic way to define getters, setters, and deleters without breaking encapsulation

14. Why is polymorphism important in OOP?
- Polymorphism is considered the key power of OOP — in fact, languages that have classes but no polymorphism are called "object-based" languages, not object-oriented languages. It makes code flexible, scalable, and clean by allowing one interface to handle many different types of objects.

15. What is an abstract class in Python?
- An abstract class is a class that cannot be instantiated directly — you can never create an object from it. It acts as a blueprint for other classes, defining methods that every child class must implement. In Python, abstract classes are created using the abc module.



16. What are the advantages of OOP?
- Advantages of OOP:
   -  Code Reusability
Write once in a parent class and reuse in all child classes through inheritance — no need to rewrite the same code again and again.
     - Modularity
Code is broken into separate classes, each handling its own job — making it easier to manage, update, and debug.
     - Data Security
Encapsulation protects sensitive data by keeping variables private and allowing access only through controlled methods.
    - Flexibility
Polymorphism lets the same method behave differently for different objects — making your code adaptable without rewriting it.
    - Scalability
Add new features by simply adding new classes without touching existing code.

17. What is the difference between a class variable and an instance variable?
- A class variable is shared across all objects of a class — one value for everyone. An instance variable is unique to each object — every object has its own copy.



18. What is multiple inheritance in Python?
- Multiple Inheritance is when a child class inherits from more than one parent class at the same time. The child class gets access to the attributes and methods of all parent classes combined.

19.  Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python.
- They Both are dunder methods that control how an object is displayed as a string — but for different audiences:

  - __str__ → For the end user — clean, readable, human-friendly output
  - __repr__ → For the developer — detailed, technical, debug-friendly output

20. What is the significance of the ‘super()’ function in Python?
- super() is a built-in function that allows a child class to call methods from its parent class — without hardcoding the parent class name. It's most commonly used inside __init__() to reuse the parent's constructor, so you don't repeat the same setup code.

21.  What is the significance of the __del__ method in Python?
- __del__ is a dunder method also known as a destructor — it is automatically called by Python's garbage collector right before an object is destroyed and removed from memory. Just like __init__ sets up an object at birth, __del__ cleans up when the object's life ends.

22. What is the difference between @staticmethod and @classmethod in Python?
- The Core Difference
  - @classmethod → receives the class itself (cls) as the first argument — can access and modify class data
​

   - @staticmethod → receives nothing automatically — no self, no cls — just a plain utility function inside a class


23.  How does polymorphism work in Python with inheritance?
- Inheritance gives the child class the parent's methods. Polymorphism lets each child class redefine those methods in their own way. Together they create code that is both reusable and flexible — one parent, many different behaviors.

24. What is method chaining in Python OOP?
- Method chaining is a technique where you call multiple methods on the same object in a single line — one after another using dots .. The key to making it work is that every method must return self — so the next method can immediately act on the same object.

25.  What is the purpose of the __call__ method in Python?
- __call__ is a dunder method that makes an object behave like a function — meaning you can call an object using parentheses () just like you call a regular function. When you define __call__ inside a class, calling obj() automatically triggers obj.__call__() behind the scenes.

```python
#1 Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".
class Animal:
    def speak(self):
        print("Animal makes a sound!")

class Dog(Animal):
    def speak(self):
        print("Bark!")


a = Animal()
d = Dog()

a.speak()
d.speak()
```

```python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w
    def area(self):
        return self.l * self.w


c = Circle(5)
r = Rectangle(4, 6)

print(c.area())
print(r.area())
```

```python
#3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

    def show(self):
        print(f"Type: {self.type} | Brand: {self.brand} | Battery: {self.battery}kWh")


ec = ElectricCar("Electric", "Tesla", 100)
ec.show()
```

```python
#4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.
class Bird:
    def fly(self):
        print("Birds can fly!")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high!")

class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly!")


birds = [Sparrow(), Penguin()]

for bird in birds:
    bird.fly()

```

```python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance        # private attribute

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited  {amount}")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance -= amount
            print(f"Withdrawn  {amount}")

    def check_balance(self):
        print(f"Balance:  {self.__balance}")


acc = BankAccount(1000)

acc.check_balance()
acc.deposit(500)
acc.check_balance()
acc.withdraw(2000)
acc.withdraw(300)
acc.check_balance()
```

```python
#6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().
class Instrument:
    def play(self):
        print("Instrument is playing...")

class Guitar(Instrument):
    def play(self):
        print("Guitar: Strumming strings! ")

class Piano(Instrument):
    def play(self):
        print("Piano: Playing keys! ")


instruments = [Guitar(), Piano()]

for ins in instruments:
    ins.play()

```

```python
#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers
class MathOperations:

    @classmethod
    def add_numbers(cls, a, b):
        print(f"Addition: {a + b}")

    @staticmethod
    def subtract_numbers(a, b):
        print(f"Subtraction: {a - b}")


MathOperations.add_numbers(10, 5)
MathOperations.subtract_numbers(10, 5)
```

```python
#9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"


f1 = Fraction(3, 4)
f2 = Fraction(7, 2)

print(f1)
print(f2)
```

```python
#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"


v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2

print(v3)
```

```python
#12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        avg = sum(self.grades) / len(self.grades)
        print(f"{self.name}'s Average Grade: {avg:.2f}")


s1 = Student("Ravi", [85, 90, 78, 92, 88])
s2 = Student("Anita", [70, 65, 80, 75])

s1.average_grade()
s2.average_grade()
```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary
class Employee:
    def __init__(self, name, hours, rate):
        self.name = name
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate


class Manager(Employee):
    def __init__(self, name, hours, rate, bonus):
        super().__init__(name, hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus


e = Employee("Ravi", 40, 200)
m = Manager("Anita", 40, 500, 5000)

print(f"{e.name}'s Salary: ₹{e.calculate_salary()}")
print(f"{m.name}'s Salary: ₹{m.calculate_salary()}")
```

```python
#16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Cow says: Moo! ")

class Sheep(Animal):
    def sound(self):
        print("Sheep says: Baa! ")


c = Cow()
s = Sheep()

c.sound()
s.sound()
```

```python
#17.  Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author} ({self.year_published})"


b1 = Book("The Alchemist", "Paulo Coelho", 1988)
b2 = Book("Atomic Habits", "James Clear", 2018)

print(b1.get_book_info())
print(b2.get_book_info())
```

```python
#18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def show(self):
        print(f"Address: {self.address} | Price: ₹{self.price}")


class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def show(self):
        print(f"Address: {self.address} | Price: ₹{self.price} | Rooms: {self.number_of_rooms}")


h = House("MG Road, Vijayawada", 5000000)
m = Mansion("Banjara Hills, Hyderabad", 50000000, 12)

h.show()
m.show()
```
